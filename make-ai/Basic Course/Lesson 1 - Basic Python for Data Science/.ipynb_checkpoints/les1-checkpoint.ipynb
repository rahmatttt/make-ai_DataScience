# 9. While Loop 

Juga merupakan salah satu metode looping, namun berbeda dengan for

```python
for i in range(1,6):
    print(i)
```

```python
i = 1
while i<=5:
    print(i)
    i = i+1
```

pastikan ada inisiasi di awal (i = 1) dan ada incrementnya (i = i+1). Kesalahan yang biasa terjadi adalah lupa inisiasi atau lupa increment. DAN JANGAN LUPA INDENTASI!

```python
#lupa inisiasi
while j<5:
    print(i)
    j = j+1
```

```python
#lupa increment
j = 1
while j < 5:
    print(j)
```

seperti yang anda lihat, itu adalah looping forever. Kayak sayangku ke kamu, forever <3

mungkin anda akan bingung lalu kapan harus pake while, berikut adalah contoh looping yang hanya bisa dilakukan dengan while silakan anda telaah sendiri

```python
cek = True
while(cek):
    print(cek)
    question = input("lanjut?")
    if question == "gamao":
        cek = False
```

```python
def plus1(x,y):
    return x+y
```

```python
plus1(10,5)
```

```python
plus1(10)
```

nah kan error, soalnya gak lengkap :(

parameter dalam function juga bisa diberi default value, contohnya kayak gini

```python
def plus2(x,y = 20):
    return x+y
```

```python
plus2(10)
```

```python
plus2(10,2)
```

```python
def plus3(x = 10, y = 20):
    return x+y
```

```python
plus3()
```

```python
plus3(1)
```

```python
plus3(1,2)
```

tapi parameter yang bisa dikasih default value harus ditaro di kanan, pokoknya jangan sampe ada parameter yang di kiri ada default valuenya tapi di kanan gak ada, nanti jadinya error kayak gini

```python
def plus4(x=10, y):
    return x+y
```

# 11. Class 

```python
class Operation:
    #concstructor
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def plus(self):
        return self.x + self.y
    def minus(self):
        return self.x - self.y
```

```python
opr = Operation(2,3)
```

```python
opr.plus()
```

```python
opr.minus()
```

```python
opr.x
```

```python
opr.y
```
