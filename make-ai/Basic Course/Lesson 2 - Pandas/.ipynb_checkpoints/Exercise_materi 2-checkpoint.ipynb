import pandas package

```python
import pandas as pd
```

input data yang bernama data_penjualan_motor_2018.xlsx dan tampilkan 5 data teratas dengan fungsi head()

```python
motor = pd.read_excel('Exercise_Python for Spreadsheet/data/data_penjualan_motor2018.xlsx')
motor.head()
```

tampilkan jumlah dari seluruh variabel numerik pada data dengan fungsi sum()

```python
motor.sum(numeric_only=True)
```

tampilkan jumlah penjualan berdasarkan Bulan dengan fungsi groupby('nama kolom')

```python
motor.groupby('Bulan', as_index = False).sum()
```

tampilkan jumlah penjualan tiap Bulan dan jenis

```python
motor.groupby(['Bulan','Jenis'], as_index=True).sum()
```

lihat rata-rata penjualan dengan fungsi mean()

```python
motor.mean()
```

```python
motor.groupby(['Bulan'], as_index=False).mean()
```

```python
motor.groupby(['Jenis'], as_index=False).mean()
```

mengurutkan data dari tabel motor dari besar ke kecil dengan bulan dan jenis

```python
totals = motor.sort_values('Jumlah', ascending=False).reset_index(drop=True)
totals
```

tampilkan data Yamaha

```python
motor1 = motor[motor['Jenis'] == 'Yamaha'].reset_index(drop=True)
motor1
```

buat fungsi get_stats() untuk melihat nilai maksimum, rata-rata, dan minimum berdasarkan jenis. Lalu terapkan fungsi tersebut dengan fungsi apply, dan jadikan bentuknya menjadi dataframe dengan fungsi unstack().

```python
def get_stats(group):
    return {'max':group.max(), 'mean':group.mean(), 'min':group.min()}

conclusion = motor['Jumlah'].groupby(motor['Jenis']).apply(get_stats).unstack()
conclusion
```
