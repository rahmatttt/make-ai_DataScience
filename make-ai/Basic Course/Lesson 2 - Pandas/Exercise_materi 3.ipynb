**Atribute**

    - format pemanggilan : object.atribute
    - contoh : workbook.sheet_names
    - menampilkan sesuatu (contoh : excel yang dipanggil)
    - selalu di-attach dalam object

**Method**

    - format pemanggilan : object.method()
    - contoh : workbook.parse()
    - melakukan sesuatu (contoh : menampilkan data untuk variabel tertentu)
    - selalu di-attach dalam object

**Function**

    - format pemanggilan : package.function()
    - contoh : pd.ExelFile()

Ubah variabel Kelamin menjadi data frame dengan parsing sheet. parse('nama_sheet')

kita harus mengetahui bahwa python adalah bahasa pemrograma yang case sensitive.

contoh :

    upper and lower case sensitive
        - 'Andi' != 'andi'
    
    space sensitive
        - 'Andi' != 'Andi '
    
    Merge table
        - AB + ACD = ABCD
        - Jika kita hanya ingin kolom ABC maka kita bisa menghapus kolom D


#### Cara mengatasi case sensitive 

dalam **data_kelamin** kita dapat lihat bahwa kita menginput data untuk variabel Nama dengan huruf kapital, kita dapat menggantinya sesuka kita.

ubah data menjadi lower letter

```python
data_kelamin['Nama'] = data_kelamin['Nama'].str.lower()
data_kelamin
```

Ubah menjadi kapital semua

```python
data_kelamin['Nama'] = data_kelamin['Nama'].str.upper()
data_kelamin
```

gunakan str.title() untuk membuat huruf pertama kapital

```python
data_kelamin['Nama'] = data_kelamin['Nama'].str.title()
data_kelamin
```

#### Space handling 

gunakan str.strip() untuk menghapus spasi di kata pertama dan kata terakhir

```python
data_kelamin['Nama'] = data_kelamin['Nama'].str.strip()
print(data_kelamin)
```

#### Memilih dan menghapus kolom 

Memilih data dari sheet Kelamin dari variabel Kelamin dan Nama

```python
data_kelamin[['Nama','Kelamin']]
```

Menghapus kolom dengan drop()

hapus kolom Kelamin

```python
data_kelamin.drop('Kelamin', axis = 1)
```

#### Merge table : VLOOKUP dalam python

Cukup mudah untuk menggabungkan **data_kelamin** dan **data_nilai** dengan variabel **Nama** menggunakan fungsi .merge(nama_sheet, on="key_column", how='join type(left/right/inner/outer)').

jika anda mengerti SQL mungkin fungsi tersebut dapat dengan mudah anda pahami, kalo gangerti belajar jangan main mulu

```python
data_kelamin.merge(data_nilai, on="Nama", how="left")
```

Contoh lain kita akan menggunakan file **data_nilai_matematika2.xlsx**

```python
data_workbook = pd.ExcelFile('Exercise_Python for Spreadsheet/data/data_nilai_matematika2.xlsx')
```

Lihat sheet Kelamin

```python
data_kelamin = data_workbook.parse('Kelamin')
data_kelamin
```

Lihat sheet Nilai

```python
data_nilai = data_workbook.parse('Nilai')
data_nilai
```

Sekarang kita akan menggabungkan dua tabel tersebut. Bedanya dengan sebelumnya adalah kolom yang akan kita gunakan memiliki nama yang berbeda, yaitu Nama dan Nama Siswa. maka dari itu kita gunakan fungsi :

.merge(data_nilai, left_on='Nama', right_on='Nama Siswa', how='left')

```python
data_gabung = data_kelamin.merge(data_nilai, left_on='Nama', right_on='Nama Siswa', how='left')
data_gabung
```

Sekarang data sudah berhasil digabungkan, namun kita lihat kolom Nama dan Nama Siswa sebenarnya sama sehingga kita akan menghilangkan salah satu kolom tersebut.

drop kolom Nama Siswa

```python
data_gabung.drop('Nama Siswa', axis=1)
```
