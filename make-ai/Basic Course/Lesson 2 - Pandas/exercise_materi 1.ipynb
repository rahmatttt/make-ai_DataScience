Sorting data berdasarkan kolom Nama

```python
data = data.sort_values('Nama')
data
```

Data berhasil terurut berdasarkan nama namun indexnya jadi acak-acakan, maka dari itu kita harus set ulang indexnya

set index dari data dengan kolom Nama yang disorting

```python
data = data.reset_index(drop=True)
data
```

Dapat kita lihat bahwa indexnya sekarang sudah sesuai

Sekarang coba sorting berdasarkan Nilai

```python
data = data.sort_values('Nilai', ascending=False) #descending
data = data.reset_index(drop=True)
data
```

Operator-operator untuk komparasi

    - == : sama dengan
    - != : tidak sama dengan

    - <= : kurang dari sama dengan
    - >= : lebih dari sama dengan

##### Create new column 

Operator matematika :
    - + (plus)
    - - (minus)
    - * (kali)
    - / (bagi)
    
buat kolom baru dengan nama Nilai_akhir dari fungsi variael Nilai+10. Artinya adalah pada kolom baru tersebut semua nilai akan ditambahkan 10

```python
data['Nilai_akhir'] = data['Nilai'] + 10
data
```

##### Create new row

buat baris baru seperti ini:
{Nama : "Tatang", Kelamin : "Laki-laki", "Nilai" : 89, "Nilai_akhir" : 89+10}

```python
data1 = [{"Nama" : "Tatang", "Kelamin" : "Laki-laki", "Nilai" : 89, "Nilai_akhir" : 89+10}]
data1
```

jadikan data frame

```python
baru = pd.DataFrame(data1)
baru
```

sekarang gunakan append()

```python
data = data.append(baru).reset_index(drop=True)
data
```
