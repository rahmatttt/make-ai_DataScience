## Import data 

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('athlete_events.csv')
df
```

data di atas adalah data atlet pada olimpiade. Kita akan membuat visualisasi data untuk data olimpiade Brazil tahun 2016. Kita akan melihat berdasarkan perolehan medali, jika kita lihat pada data di atas pada kolom Medal banyak ditemukan NaN yang artinya value nya tidak ada, maka dari itu kita harus membuatnya memiliki value

## Mengatasi missing value

Kita cek dulu tahun berlangsungnya olimpiade

```python
df['Year'].unique()
```

Setelah itu kita cek ukuran dari data

```python
df.shape
```

kita ambil data tahun 2016

```python
df = df[df['Year'] == 2016]
df
```

cek ukurannya lagi

```python
df.shape
```

ada 13688 data dari tahun 2016

sekarang isi value yang kosong dengan 'none' dengan fungsi fillna(). fungsi fillna() adalah fungsi yang berfungsi untuk mengisi field yang kosong

```python
df = df.fillna({'Medal' : 'none'})
```

```python
df.head()
```

data NaN sudah berubah jadi none

## Mengubah data kategorikal menjadi numerikal 

sekarang kita akan mencari tau berapa medali yang diperoleh tiap negara

```python
All_go = pd.DataFrame(df.groupby(['NOC','Medal'])['Medal'].size())
All_go
```

Data di atas bentuknya adalah multi indeks, maka dari itu kita harus mengubahnya menjadi satu indeks. Namun sebelum kita buat jadi satu indeks kita lihat bahwa ada dua kolom Medal disitu, maka dari itu kita harus mengubah salah satunya

```python
All_go.columns = ['Total']
All_go
```

```python
All_go = All_go.reset_index()
All_go.head()
```

Sekarang data tersebut sudah memiliki satu index namun jika kita lihat pada index ke 2 dan 3 data tersebut adalah sama (sama-sama ALG) walaupun Medalnya berbeda

```python
All_go['None'] = np.where(All_go['Medal'] == 'none', All_go['Total'], 0)
All_go['G'] = np.where(All_go['Medal'] == 'Gold', All_go['Total'], 0)
All_go['S'] = np.where(All_go['Medal'] == 'Silver', All_go['Total'], 0)
All_go['B'] = np.where(All_go['Medal'] == 'Bronze', All_go['Total'], 0)
```

```python
All_go.head()
```

```python
All_go = All_go.groupby(['NOC'], as_index = False).sum()
```

```python
All_go
```

Mari kita sorting berdasarkan medalinya

```python
All_go = All_go.sort_values(['G','S','B'], ascending=False)
```

```python
All_go
```

Jika kita perhatikan, kolom Total di atas adalah total pertandingan yang diikuti, sekarang kita akan mencari tahu berapa total medali tiap negara

    Total_medal = Total - None

```python
All_go['Total_medal'] = All_go['Total'] - All_go['None']
```

```python
All_go
```

# Line Chart

Line chart bagus digunakan untuk melihat trend. Sekarang kita akan melihat trend perolehan medali dari tahun ke tahun untuk negara tertentu

```python
df1 = pd.read_csv('athlete_events.csv')
df1
```

```python
df1 = df1.fillna({'Medal' : 'none'})
```

```python
df1.head()
```

```python
All_new = pd.DataFrame(df1.groupby(['Year','NOC','Medal'])['Medal'].size())
All_new
```

```python
All_new.columns = ['Total']
All_new.head()
```

```python
All_new = All_new.reset_index()
```

```python
All_new.head()
```

```python
All_new['None'] = np.where(All_new['Medal'] == 'none', All_new['Total'], 0)
All_new['G'] = np.where(All_new['Medal'] == 'Gold', All_new['Total'], 0)
All_new['S'] = np.where(All_new['Medal'] == 'Silver', All_new['Total'], 0)
All_new['B'] = np.where(All_new['Medal'] == 'Bronze', All_new['Total'], 0)

All_new = All_new.groupby(['Year','NOC'], as_index = False).sum()
```

```python
All_new
```

```python
All_new['Total_medal'] = All_new['Total'] - All_new['None']
```

```python
All_new.head()
```

kita ambil data GRE (Yunani)

```python
gre = All_new[All_new['NOC'] == 'GRE']
gre
```

kita gunakan seaborn

```python
sns.lineplot(x='Year', y='Total_medal', data=gre)
```

# Multi Line Chart

```python
gre
```

```python
gre = gre.sort_values(['Year'])
```

```python
plt.figure(figsize=(20,12))
plt.plot(gre['Year'], gre['G'], marker='o', markersize =12, color = 'gold', linewidth = 4, label='gold')
plt.plot(gre['Year'], gre['S'], marker='+', markersize =12, color = 'silver', linewidth = 4, label='silver')
plt.plot(gre['Year'], gre['B'], marker = 'x', markersize =12, color = 'brown', linewidth=4, linestyle='dashed', label='bronze')
plt.legend()
plt.title("trend perolehan medali Yunani")
plt.show()
```

# Statistical Chart

### Box Plot 

Box Plot cocok untuk data yang bertipe numerical dan categorical untuk mengatahui persebaran dan outlier data yang kita gunakan. 

Pada kasus ini Box Plot dibuat dari hubungan variabel Sex dan Age, hubungan variabel Sex dan Height, dan hubungan variabel Sex dan Weight. 

```python
df.head()
```

```python
sns.boxplot(x = 'Sex', y = 'Age', data=df)
plt.title('Persebaran umur berdasarkan jenis kelamin')
plt.show()
```

Dari boxplot diatas didapatkan banyak data yang outlier baik dari perempuan maupun laki-laki. Persebaran umur perempuan dan laki-laki hampir sama, hanya saja perempuan terdapat beberapa atlet yang lebih tua dari pada perempuan. 

```python
sns.boxplot(x = 'Sex', y = 'Height', data=df)
plt.title('Persebaran tinggi berdasarkan jenis kelamin')
plt.show()
```

Dari boxplot diatas diketahui bahwa kecendrungan data berdasarkan tinggi. Pada boxplot laki-laki persebaran tidak berbeda jauh dari pada perempuan. Tetapi, laki-laki persebaran data sedikit lebih tinggi. 

```python
sns.boxplot(x = 'Sex', y = 'Weight', data=df)
plt.title('Persebaran berat berdasarkan jenis kelamin')
plt.show()
```

Didapatkan kecendrungan persebaran data laki-laki sedikit lebih tiggi dari pada perempuan. Outlier pada perempuan ada outlier dibawah nilai minimal sedangkan laki-laki tidak ada.

### Histogram 

Histogram visualisasinya hampir sama dengan Bar Chart dari pembahasan sebelumnya. Tetapi, Bar Chart digunakan untuk data yang bertipe categorical sedangkan Histogram untuk data yang bertipe Numerical. 

Perbedaan yang lain yaitu pada Bar Chart visualisasi antar batang akan ada spasinya atau jaraknya. Sedangkan, pda Histogram visualisasi antar batang akan berdempetan. 


```python
sns.distplot(df['Age'], kde=False)#kde berhubungan dengan distribusi normal
plt.ylabel('count')
plt.title('distribution of age')
plt.show()
```

Dari histogram diatas dapat didapatkan informasi atlet yang paling banyak mengikuti olympiade yaitu dengan umur 20-30 tahun.

### Heading 

Heatmap digunakan untuk mengatahui korelasi antar data dengan warna yang menarik. Nah, pada Heatmap ini selain kita bisa melihat besar korelasi dengan menggunakan angka, kita bisa juga melihat korelasi melalui warna. Korelasi nilanya yaitu -1 sampai 1. Ketika nilai mendekati 1 (korelasi positif) dan -1 (korelasi negatif) maka nilai tersebut berkorelasi kuat. 

```python
df.corr()
```

sebenarnya mencari korelasi akan lebih baik bila datanya berupa angka semua, maka kita akan gunakan All_go

```python
All_go.corr()
```

sekarang kita akan mengubahnya dalam bentuk heatmap

```python
sns.heatmap(All_go.corr())
plt.show()
```

jika kita ingin heatmap tetap menampilkan angka maka lakukan seperti ini

```python
sns.heatmap(All_go.corr(), annot=True)
plt.show()
```

kita bisa merubah warnanya, untuk jenis warnanya dapat dilihat di https://matplotlib.org/3.1.1/tutorials/colors/colormaps.html

```python
sns.heatmap(All_go.corr(), annot=True, cmap='spring')
plt.show()
```
