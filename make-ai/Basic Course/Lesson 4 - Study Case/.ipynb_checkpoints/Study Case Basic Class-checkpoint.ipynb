let's sorting **data_match** based on **Team** and **Date**

```python
data_match = data_match.sort_values(by=['Team', 'Date'])
```

because there 128 rows, we are trying to use only match in **Group Stage** , so we should filter our data based on **Round** column, select only **Group Stage**

```python
data_match = data_match[data_match['Round'] == 'Group Stage']
```

```python
data_match
```

Check data that **Goal Scored** is more than 3

```python
data_match[data_match['Goal Scored'] > 3]
```

Now let's create new column, called **Passes Completed**, what's that?

**Passes Completed** is **Pass Accuracy** times **Passes**. So, how to do that?

```python
data_match['Passes Completed'] = (data_match['Pass Accuracy %']/100) * data_match['Passes']
```

```python
data_match.head()
```

Check statistic value (mean of data) based on how many **Goal Scored**

```python
data_match.groupby(['Goal Scored']).mean()
```

Let's check based on **Team** and **Goal**

```python
pd.DataFrame(data_match.groupby(['Team'])['Goal Scored'].mean())
```

Let's check so that all of data is visible and doing some groupby() function

```python
data_match.groupby(['Team','Goal Scored']).mean()
```

Get data with at least **Goal Scored** is 2 and **Passes Complete** more than 500

```python
data_match[(data_match['Goal Scored'] == 2) & (data_match['Passes Completed'] > 500.0)]
```

menggunakan viz_1

```python
viz_1 = viz_1.sort_values(['Goal Scored'], ascending=False)
viz_1
```

```python
plt.figure(figsize=(20,12))
plt.barh(viz_1['Team'], viz_1['Goal Scored'], color="r")
plt.title("Total Goals in Group Stage FIFA World Cup Russia 2018")
plt.xlabel('Total Goals')
plt.ylabel('Team')
plt.show()
```

**Data Kedua**

melihat kecenderungan antara Pass Accuracy dan Passes Complete.Dalam hal ini kita gunakan rata-rata Pass Accuracy dan rata-rata Passes Complete dari masing-masing team

```python
viz_2 = data_match[['Team', 'Pass Accuracy %','Passes Completed']]
```

```python
viz_2
```

```python
viz_2 = viz_2.groupby(['Team']).mean()
```

```python
viz_2
```

```python
sns.scatterplot(x='Pass Accuracy %',y='Passes Completed', data=viz_2)
plt.title('trend between Pass Accuracy and Passes Completed')
plt.show()
```

**Data Ketiga**

Pada visualisasi ketiga ini saya ingin melihat distribusi data Attempts, sehingga saya mengetahui dalam satu pertandingan rata-rata team dapat melakukan berapa kali percobaan shoot.

```python
sns.distplot(data_match['Attempts'], kde=False)
plt.ylabel('count')
plt.title('distribution of attempts')
plt.show()
```

##### Penjelasan visualisasi 

**1. Data pertama**

grafik ini adalah grafik batang, yang menunjukkan total goal dari tiap team pada group stage FIFA World Cup Russia 2018. Dari visualisasi data tersebut kita dapat mengetahui bahwa team yang memiliki jumlah goal terbanyak dalam group stage adalah Belgium dengan total goal lebih dari 8.

**2. Data kedua**

grafik adalah grafik scatter, yang menunjukkan kecenderungan antara pass accuracy dan passes complete pada group stage FIFA World Cup Russia 2018. Dari visualisasi data tersebut kita dapat mengetahui bahwa pass accuracy dan passes complete memiliki kecenderungan positif, dimana semakin besar pass accuracy maka semakin besar passes complete, begitu juga sebaliknya.

**3. Data ketiga**

grafik ini adalah histogram, yang menunjukkan distribusi dari data Attempts pada group stage FIFA World Cup Russia 2018. Dari visualisasi data tersebut kita dapat mengetahui bahwa dalam suatu pertandingan rata-rata team akan melakukan percobaan goal atau attempts sebanyak 10 sampai 15 kali.
