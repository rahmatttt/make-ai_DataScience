Inferential statistics = memprediksi apa yang akan terjadi selanjutnya berdasarkan data yang sudah ada

descriptive statistics:

    - measure of central tendency

### Measure of Central Tendency 

Mean, Median, Modus

Mean = rata-rata
Median = nilai tengah
Modus = nilai yang paling sering keluar

```python
data2 = pd.DataFrame({'e':[5,3,4,5,6,9,10,3,4,3]})
data2
```

```python
print("Mean : ", data2['e'].mean())
print("Median : ", data2['e'].median())
print("Modus : ", data2['e'].mode().values[0]) #bentuk modus memang berbeda dari yang lain btw kalo di sini namanya mode
```

**Mean** kurang cocok dijadikan central tendency jika ada anomali pada data. **Median** bersifat lebih stabil sehingga cocok untuk digunakan ketika ada anomali. **Modul** digunakan ketika datanya berbentuk bukan angka.

contoh:

```python
data2 = pd.DataFrame({'e':[5,3,4,5,6,9,10,3,4,3,100]})
data2
```

```python
print("Mean : ", data2['e'].mean())
print("Median : ", data2['e'].median())
print("Modus : ", data2['e'].mode().values[0]) #bentuk modus memang berbeda dari yang lain btw kalo di sini namanya mode
```

dari output di atas dapat kita lihat bahwa setelah ditambahkan data 100, mean berubah dengan sangat signifikan. Itu menunjukkan bahwa mean kurang cocok digunakan ketika ada anomali pada data. 100 itu sendiri disebut sebagai **outlier**

### Measure of Dispersion 

Measure of Dispersion adalah bagaimana cara kita mencaritahu persebaran data. Persebaran data dapat diketahui melalui range. Dengan mempelajari measurement of Dispersion, kita akan mengetahui persebaran data, yakni bagaimana nilai-nilai dari sebuah data tersebut tersebar. Apakah ia berkelompok, sehingga datanya mengumpul pada nilai tertentu, ataukah justru nilainya tersebar sehingga nilai variansinya tinggi

    Range = nilai max - nilai min

Gambar di atas adalah perbandingan data dengan variansi yang besar dan dengan variansi yang lebih kecil. Ketika kita memiliki variansi yang besar maka rangenya juga akan semakin besar. Data sebelah kiri memiliki range 45 sedangkan sebelah kanan adalah 17.

Fungsi standart deviasi adalah biasanya digunakan jika ada hubungan dengan rata-rata. Dengan mengetahui rata-rata dan standart deviasi maka kita bisa tahu data tersebut ada di range berapa hingga berapa.

**Data sebelah kiri**

untuk mengecek hubungan antara mean dan standart deviasi:
    1. mean - standart deviasi = kira kira 13
    2. mean + standart deviasi = kira kira 41

jika kita lihat ada 6 dari 10 data di antara 13 hingga 41 yang membuktikan bahwa mayoritas data tersebut berada di antara 13 - 41.

**variansi = power(standart deviasi,2)**

kita gunakan data2

standart deviasi

```python
data2['e'].std()
```

mungkin agak sedikit aneh ketika rata-rata kurang dari standar deviasi, tapi jangan lupa bahwa data2 memiliki outlier

variansi

```python
data2['e'].var()
```

variansinya cukup besar, begitu pula rangenya (100-3 = 97) artinya perseberannya juga besar (disebabkan ada outlier)

### Measure of Position 

measurement of position adalah pengukuran interquartil. Dengan mengetahui measurement of position ini, kita akan lebih mudah dalam menyimpulkan distribusi dari sebuah data. Interquartil ini akan kemudian lebih mudah apabila divisualisasikan dalam bentuk boxplot.

**Quartil :**
    - Q1 : quartil 1 (25%)
    - Q2 : quartil 2/median (50%)
    - Q3 : quartil 3(75%)
    - Q3>Q2>Q1

**Interquartil :** Q3 - Q1

**Batas minimum :** Q1 - (1.5 x IQR)

**Batas maksimum :** Q3 + (1.5 x IQR)

data yang kurang dari batas minimum atau lebih dari batas maksimum disebut outlier (pencilan)


Kita gunakan data2 untuk mencaritahu nilai quartil

```python
data2.describe()
```

fungsi describe adalah fungsi untuk menunjukkan data statistik dari sebuah data.

**keterangan :**
    - count : banyak data
    - mean : rata-rata
    - std : standar deviasi
    - min : nilai terendah
    - 25% : Q1
    - 50% : Q2(median)
    - 75% : Q3
    - max : nilai tertinggi

sekarang kita coba hitung interquartil, batas minimum, dan maksimumnya

```python
#interquartil
IQR = 7.5 - 3.5
IQR
```

```python
#batas minimum
3.5 - (1.5*IQR)
```

```python
#batas maksimum
7.5 + (1.5*IQR)
```

dari hasil di atas dapat diketahui bahwa pada data2, nilai 100 adalah outlier

sekarang kita gunakan boxplot

```python
plt.figure(figsize=(12,5))
sns.boxplot(x = 'e', data=data2)
plt.title('Persebaran nilai pada data2')
plt.show()
```

### Inferential Statistics (Correlation)

Inferensial statistik umumnya digunakan ketika kita hendak membuat sebuah hipotesis, prediksi maupun membuat kesimpulan.

Nilai korelasi digunakan untuk mengetahui hubungan antar variabel. Pada pandas dataframe kita dapat dengan mudah mendapatkan nilai correlation dengan menggunakan perintah df.corr(). Umumnya korelasi pada pandas menggunakan metode penghitungan Pearson, namun kita juga dapat secara eksplisit mengganti dengan metode penghitungan Spearman.

Korelasi positif : berbanding lurus

korelasi negatif : berbanding terbalik

grafik di atas adalah contoh grafik yang menunjukkan korelasi antara X dan Y dengan r adalah nilai korelasi. Jika r bernilai 0.0 menunjukkan bahwa tidak ada hubungan antara X dan Y, sedangkan jika X mendekati satu menunjukkan bahwa ada korelasi positif antara X dan Y, jika mendekati -1 maka ada korelasi negatif antara X dan Y. Korelasi bernilai dari -1 hingga 1.

**Pearson VS Spearman**

pearson
    korelasi barusan adalah pearson. Pearson ideal untuk menghitung korelasi antara data kontinu
    
spearman
    digunakan untuk data ordinal (jenis kelamin (1 untuk laki, 0 untuk perempuan)) tapi ada beberapa kondisi untuk kontinu juga

```python
data.corr(method='spearman')
```

Lihat bagian korelasi SibSp(Sib Spouse) dan Fare, ada perbedaan yang cukup signifikan dimana spearman jauh lebih besar dikarenakan SibSp adalah ordinal.

**Kapan gunakan pearson?**

    Ketika seluruh datanya adalah kontinu dan tidak ada data ordinal

**spearman?**
    
    Ketika ditemukan data ordinal

### Inferential Statistics (Normalization)

#### Distribusi Normal

Bentuk data tidak semuanya sempurna (tidak berdistribusi dengan baik/normal). Kebanyakan data-data yang didapatkan adalah negatif skew (nilai datanya kebanyakan besar) atau positif skew (nilai datanya kebanyakan kecil). Maka dari itu muncul kondisi dimana kita akan menormalisasi data agar gap atau jarak antar value tidak terlalu jauh dan untuk meminimalisir outlier. Normalisasi data menggunakan sesuatu yang disebut dengan **Z Score**

**keterangan**

- Score = Nilai data
- Mean = rata-rata
- SD = Standar deviasi (mencari range mayoritas data)

Let's Practice!

```python
a = [26,33,65,28,34,55,25,44,50,36,26,37,43,62,35]
df = pd.DataFrame({'distribution' : a})
df
```

sekarang kita ingin menormalkan data di atas. Berdasar rumus kita perlu mean dan standar deviasi, maka dari itu kita akan gunakan fungsi mean() dan std()

```python
mean = df['distribution'].mean()
print('Mean = ', mean)
```

```python
std = df['distribution'].std()
print('Standar deviasi = ', std)
```

Hubungan antara mean dan std adalah data berada pada **mean-std** hingga **mean+std**, dari hitungan di atas menunjukkan bahwa mayoritas data kita ada di antara **27 hingga 53**.


Sekarang kita akan lakukan normalisasi, coba buat kolom baru bernama **normalize**

```python
df['normalize'] = round((df['distribution']-mean)/std,2)
#round adalah fungsi pembulatan dimana parameter kedua adalah jumlah angka di belakang koma
df
```

jika kita lihat hasil normalisasi rangenya tidak terlalu jauh. Itulah fungsi dari normalisasi untuk menormalkan data
