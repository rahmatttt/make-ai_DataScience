# Linear Model

Linear model adalah metode statistik yang berfungsi untuk mengetahui hubungan antara satu fitur dengan fitur lainnya.

persamaan linear dapat ditulis dalam bentuk :
    y = mx + b
    
    ket :
    - y = variabel target atau hasil / depended (akibat)
    - x = faktor independen (sebab)
    
contoh :
    tinggi_badan = 2 * berat_badan + 50

### Contoh model linear

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Misalnya, kita mempunyai data data berat badan dan tinggi badan. Kita dapat membentuk garis regresi linier hubungan antara berat badan dan tinggi badan pada data yang akan kita inputkan berikut

```python
dt = pd.DataFrame({'Height':[150,160,170,175,165,155, 180],
                   'Weight':[40,50,60,65,55,45,90]})
sns.lmplot(x='Height', y='Weight', data=dt, ci=False)
plt.show()
```

Dari gambar diatas didapatkan garis regresi linier yang artinya ketika tinggi badan 150 maka berat badan 40, ketika tinggi 155 maka berat badan 45, dst.

### How to Get Function of Linear Model

#### y = mx + b

ket :
- y = variabel target atau variabel terikat/dependent
- x = variabel fitur atau variabel bebas/independent
- m = gradient atau slope (garis miring)
- b = intercept (pergeseran data)

Dari rumus diatas merupakan perhitungan yang disebut fitting data.

### Menghitung Slope

kita misalkan
- a = n(sigma(xy))
- b = (sigma(x))(sigma(y))
- c = n(sigma(x * x))
- d = sigma(x)^2

```python
dt
```

```python
a = len(dt)*((dt['Height']*dt['Weight']).sum())
a
```

```python
b = (dt['Height'].sum())*(dt['Weight'].sum())
b
```

```python
c = len(dt)*((dt['Height']**2).sum())
c
```

```python
d = (dt['Height'].sum())**2
d
```

```python
m = (a-b)/(c-d)
m
```

hasil m adalah slope nya

### Intercept

kita misalkan :
- e = (sigma(y))(sigma(x^2))
- f = (sigma(x))(sigma(xy))

```python
e = (dt['Weight'].sum())*((dt['Height']**2).sum())
e
```

```python
f = (dt['Height'].sum())*((dt['Height']*dt['Weight']).sum())
f
```

```python
bb = (e-f)/(c-d)
bb
```

Didapatkan besarnya nilai intercept sebesar -177.85714285714286.

```python
print('Weight =', m, '* Height +',bb)
```

Arti dari output diatas yaitu ketika kita mempunyai nilai tinggi badan maka kita kalikan terlebih dahulu 1.4285714285714286 setelah itu ditambah dengan -177.85714285714286, maka kita dapat nilai untuk berat badan dari formula diatas.

```python
reg.intercept_
```

## Regression Metrics

ada beberapa tipe metrik (pengukuran/pengecekan akurasi) dalam regresi, namun yang paling umum digunakan ada 3 :
- Mean Square Error (MSE)
- Mean Absolute Error (MAE)
- R-Square (R2)

**MSE**

merupakan suatu pengukuran dengan merata-ratakan nilai-nilai error yang dikuadratkan.

**MAE**

merupakan suatu pengukuran dengan mearta-ratakan nilai error yang diabsolutkan

**R-Square**

merupakan akurasi dari analisis yang telah kita lakukan yang biasanya dibentuk dalam persentase. perhitungan dimulai satu kurangi dengan Sum Square Prediction yang telah dibagi dengan Sum Square Mean

Pada MSE dan MAE semakin kecil nilai semakin baik (karena yang dihitung adalah *error*), sedangkan R-square semakin besar nilai semakin baik (karena yang dihitung adalah *akurasi*).

### R2 Calculation

```python
hasil
```

- tanpa library

```python
sspred = (hasil['Error']**2).sum()
sspred
```

```python
weight_avg = hasil['Weight'].mean()
ssmean = ((hasil['Weight']-weight_avg)**2).sum()
ssmean
```

```python
r2_score = 1-(sspred/ssmean)
r2_score
```

- dengan library

```python
from sklearn.metrics import r2_score
```

```python
r2_score(weight_pred, y)
```

hasilnya memang berbeda tetapi tidak terlalu signifikan sehingga sebenarnya tidak masalah.
