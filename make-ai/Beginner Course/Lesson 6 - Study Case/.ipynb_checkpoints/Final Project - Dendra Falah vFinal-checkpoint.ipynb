```python
def quick_check(df):
    x = []
    for col in df.columns:
        v_c = df[col].value_counts()
        for i in range(len(v_c)):
            x.append([col,v_c.values[i], str(type(v_c.keys()[i]))])
    x = pd.DataFrame(x, columns = ['col', 'amount', 'type'])
    x = x.groupby(['col', 'type'], as_index = False).sum()
    for col in x['type'].unique():
        x[col]=col
    for col in x['type'].unique():
        x[col]=(x[col] == x['type'])*1*x['amount']
    x = x.drop(['type', 'amount'], axis = 1)
    x = x.groupby('col', as_index = False).sum()
    x['missing'] = 0
    for i in range(len(x)):
        x['missing'][i] = len(df[x['col'][i]]) - df[x['col'][i]].count()
    for j in x.drop('col', axis = 1).columns:
        x[j] = x[j]/len(df)*100
    x['unique'] = 0
    for k in range(len(x)):
        x['unique'][k] = len(df[x['col'][k]].unique())
    x['top_count'] = 0
    for l in range(len(x)):
        x['top_count'][l] = df[x['col'][l]].value_counts().values[0]/len(df)*100
    x['min_count'] = 0
    for m in range(len(x)):
        x['min_count'][m] = df[x['col'][m]].value_counts().values[len(df[x['col'][m]].value_counts().values)-1]/len(df)*100
    y = x.columns[1:4]
    x['type_amount'] = 0
    col = 0
    while col < len(y):
        x['type_amount'] += (x[y[col]] > 0.0)*1
        col += 1
    return x
```
